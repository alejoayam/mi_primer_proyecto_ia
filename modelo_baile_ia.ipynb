```python
# =========================================
# MI PRIMER MODELO DE INTELIGENCIA ARTIFICIAL
# =========================================

# 1 Importar librerías
# Las librerías son herramientas que ya existen en Python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# 2 Crear un pequeño dataset
# edad, horas_practica, le_gusta_bailar
datos = [
    [10, 5, 1],
    [12, 3, 1],
    [15, 1, 0],
    [18, 2, 0],
    [20, 6, 1],
    [25, 1, 0],
]

# Convertimos los datos en una tabla
df = pd.DataFrame(datos, columns=["edad", "horas_practica", "le_gusta"])
print("Dataset de entrenamiento")
print(df)

# 3 Variables de entrada
X = df[["edad", "horas_practica"]]

# 4 Variable que queremos predecir
y = df["le_gusta"]

# 5 Crear modelo de inteligencia artificial
modelo = DecisionTreeClassifier()

# 6 Entrenar el modelo
modelo.fit(X, y)

# 7 Realizar una predicción
edad = 16
horas = 4
prediccion = modelo.predict([[edad, horas]])

print("Predicción para persona de", edad, "años y", horas, "horas de práctica")

if prediccion[0] == 1:
    print("Le gusta bailar")
else:
    print("No le gusta bailar")
```
